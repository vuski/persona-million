```python
# [Cell 1] vote_results_all.csv → v2 + gpt-5.4-mini 페르소나 원본 컬럼만 추출
from pathlib import Path
import pandas as pd

BASE = Path.cwd()
SRC = BASE / "vote_results_all.csv"
OUT = BASE / "personas_from_votes.csv"

# 필터: v2 컨텍스트로 gpt-5.4-mini 에 보냈던 페르소나만
TARGET_MODEL = "openai/gpt-5.4-mini"
TARGET_VERSION = "v2"

PERSONA_COLS = [
    "persona_uuid",
    "sex", "age", "marital_status", "family_type", "housing_type",
    "education_level", "bachelors_field", "occupation",
    "province", "district",
    "persona_summary",
    "professional_persona", "sports_persona", "arts_persona",
    "travel_persona", "culinary_persona", "family_persona",
    "cultural_background", "skills_and_expertise",
    "hobbies_and_interests", "career_goals_and_ambitions",
    "model",
]

df = pd.read_csv(SRC, encoding="utf-8-sig")
print(f"원본: {len(df):,} rows × {len(df.columns)} cols")

mask = (df["model"] == TARGET_MODEL) & (df["voter_context_version"] == TARGET_VERSION)
sub = df[mask]
print(f"{TARGET_MODEL} × {TARGET_VERSION}: {len(sub):,} rows")

personas = sub[PERSONA_COLS].drop_duplicates(subset=["persona_uuid"]).reset_index(drop=True)
print(f"고유 페르소나: {len(personas):,} rows × {len(personas.columns)} cols")

personas.to_csv(OUT, index=False, encoding="utf-8-sig")
print(f"저장 → {OUT}")
personas.head(3)
```

```python
# [Cell 2] 100명 → 10개 청크로 분할, 각 청크별 CSV + 풀 페르소나 JSON 저장
import json

INTERESTS_DIR = BASE / "interests"
INTERESTS_DIR.mkdir(exist_ok=True)
CHUNK_SIZE = 10

target = personas.reset_index(drop=True)
print(f"전체 페르소나: {len(target):,}명")

n_chunks = (len(target) + CHUNK_SIZE - 1) // CHUNK_SIZE
PERSONA_FIELDS = [c for c in PERSONA_COLS if c != "model"]

for i in range(n_chunks):
    chunk = target.iloc[i * CHUNK_SIZE : (i + 1) * CHUNK_SIZE].reset_index(drop=True)
    # CSV
    csv_path = INTERESTS_DIR / f"personas_gpt-5.4-mini_{i+1:03d}.csv"
    chunk.to_csv(csv_path, index=False, encoding="utf-8-sig")
    # 에이전트가 읽을 풀 페르소나 JSON (idx 포함)
    out = []
    for j, r in chunk.iterrows():
        rec = {"idx": int(j)}
        for c in PERSONA_FIELDS:
            v = r[c]
            rec[c] = int(v) if c == "age" else (None if pd.isna(v) else str(v))
        out.append(rec)
    json_path = INTERESTS_DIR / f"_{i+1:03d}_personas.json"
    with open(json_path, "w", encoding="utf-8") as f:
        json.dump(out, f, ensure_ascii=False, indent=2)
    # 결과 폴더
    (INTERESTS_DIR / f"_{i+1:03d}_results").mkdir(exist_ok=True)

print(f"저장 완료: {n_chunks}개 청크 → {INTERESTS_DIR}")
print(f"마지막 청크 크기: {len(target) - (n_chunks - 1) * CHUNK_SIZE}")
```
